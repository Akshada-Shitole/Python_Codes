```python
import numpy as np
a=np.array([0,1,2,3,4,5])
print(a)
print(type(a))
```

```python
#1-D
#print dimensions
print(a.ndim)
#print number of rows and columns
print(a.shape)
#print length of array
len(a)
```

```python
#2-D
b=np.array([[0,1,2],[3,4,5]])
print(b)
```

```python
#2-D
#print dimensions
print(b.ndim)
#print number of rows and columns
print(b.shape)
#print length of array
len(b)
```

```python
#3-D
c=np.array([[[0,1],[2,3]],[[4,5],[6,7]]])
print(c)
```

```python
a=np.arange(20)
a
```

```python
#start,end(exclusive),step
b=np.arange(1,10,2)
b
```

```python
#using linspace we can divide number range in given of parts
#start,end,number of parts
a=np.linspace(0,1,5)
a
```

```python
#print 1 in 3 x 3 array
a=np.ones((3,3))
a
```

```python
#print 0 in 3 x 3 array
b=np.zeros((3,3))
b
```

```python
#eye function returns all the diagonal elements same
#returns a 2d array with ones on diagonals and zeros elsewhere
c=np.eye(2)
c
```

```python
#constructs a diagonal array
a=np.diag([1,2,3,4])
a
```

```python
#extracts the diagonal elements in an array
np.diag(a)
```

```python
#create array using random function
#create an array with given shape and populate random values
a=np.random.rand(3)
a
```
